```python
def pairplot(df, target):
    ncol, nrow = 7, df.shape[1] // 7 + (df.shape[1] % 7 > 0)
    plt.figure(figsize=(ncol * 4, nrow * 4))

    for i, feature in enumerate(df.columns):
        plt.subplot(nrow, ncol, i + 1)
        plt.scatter(df[feature], target, s=10, marker='o', alpha=.6)
        plt.xlabel(feature)
        if i % ncol == 0:
            plt.ylabel('target')
```

__Задание 2.__ Подбор гиперпараметров.

**(1 балл)**

Для Lasso- и Ridge-регресий подберите коэффициент регуляризации по обучающей выборке с помощью кросс-валидации. Параметры для перебора возьмите по логарифмической сетке от $10^{-6}$ до $10^6$).
Также посчитайте метрику $R^2$ для тестовой выборки и сравните с предыдущими результатами. Заметно ли изменилось качество?

Useful: GridSearchCV, RidgeCV, LassoCV

```python
from sklearn.linear_model import RidgeCV, LassoCV
import numpy as np

alphas = np.logspace(-6, 6, 13)

lr_with_l1reg = LassoCV(alphas=alphas)
lr_with_l1reg.fit(x_train, y_train)
print('Для L1-регуляризации выбранный параметр alpha = ', lr_with_l1reg.alpha_)

lr_with_l2reg = RidgeCV(alphas=alphas)
lr_with_l2reg.fit(x_train, y_train)
print('Для L2-регуляризации выбранный параметр alpha = ', lr_with_l2reg.alpha_)

print()
print('Новые значения метрики R2')
print('С L1-регуляризацией: ', lr_with_l1reg.score(x_test, y_test))
print('С L2-регуляризацией: ', lr_with_l2reg.score(x_test, y_test))
```

Постройте график зависимости $R^2$ для обучающей и тестовой (на кросс-валидации) выборок в зависимости от значения гиперпараметра. Для Lasso также постройте график зависимости количества ненулевых весов.

```python
scores_train = np.empty_like(alphas)
scores_test = np.empty_like(alphas)
number_nonzero = np.empty_like(alphas)

for i in range(alphas.size):
    lr = Lasso(alpha=alphas[i])
    lr.fit(x_train, y_train)
    scores_train[i] = lr.score(x_train, y_train)
    scores_test[i] = lr.score(x_test, y_test)
    number_nonzero[i] = (lr.sparse_coef_ != 0).sum()

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title('Lasso, R2', fontsize=15)
plt.semilogx(alphas, scores_train, label='Обучающая выборка')
plt.semilogx(alphas, scores_test, label='Тестовая выборка')
plt.xlabel('alpha')
plt.ylabel('R2')
plt.legend()

plt.subplot(1, 3, 2)
plt.title('Lasso, zero coefs', fontsize=15)
plt.semilogx(alphas, number_nonzero)
plt.xlabel('alpha')
plt.ylabel('Количество ненулевых весов')

for i in range(alphas.size):
    lr = Ridge(alpha=alphas[i])
    lr.fit(x_train, y_train)
    scores_train[i] = lr.score(x_train, y_train)
    scores_test[i] = lr.score(x_test, y_test)
    
plt.subplot(1, 3, 3)
plt.title('Ridge, R2', fontsize=15)
plt.semilogx(alphas, scores_train, label='Обучающая выборка')
plt.semilogx(alphas, scores_test, label='Тестовая выборка')
plt.xlabel('alpha')
plt.ylabel('R2')
plt.legend()

plt.tight_layout()
```

В какой момент происходит недообучение? Почему?

__Ответ__: При больших значениях параметров регуляризации часть функции потерь, отвечающая за регуляризацию, весит очень много, и учитывается только её вклад. Поэтому в линейной резуляризации при больших alpha (alpha > 1) все веса становятся равны нулю, то есть модель слишком простая, происходит недообучение. При этом при маленьких значениях параметра регуляризации (alpha < 0.1) обе модели на обучающей выборке показывают результаты лучше, чем на тестовой, то есть модель "запоминает" обучающую выборку, переобучается на ней.

При помощи `Pipeline` и `GridSearchSCV` выберите оптимальную архитектуру, комбинируя различные методы масштабирования, степень полинома в `PolynomialFeatures`, а также регуляризаторы. Для `Lasso` поставьте максимальное количество итераций больше значения по умолчанию, чтобы оптимизация сошлась.

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('poly', PolynomialFeatures()),
    ('clf', LinearRegression())])

params = {
    'scaler': [StandardScaler(), MinMaxScaler()],
    'poly__degree': range(1, 5),
    'clf': [Lasso(max_iter=10000), Ridge()],
    'clf__alpha': alphas,
}

grid = GridSearchCV(pipe, params)
grid = grid.fit(x_train, y_train)
```

```python
clf = grid.best_params_['clf']
scaler = grid.best_params_['scaler']
degree = grid.best_params_['poly__degree']

pipe = Pipeline([
    ('scaler', scaler),
    ('poly', PolynomialFeatures(degree)),
    ('clf', clf)])
pipe.fit(x_train, y_train)
print('R2 на подобранных параметрах модели и предобработки: ', pipe.score(x_test, y_test))
```

Для одной комбинаций метода масштабирования и линейной регрессии постройте зависимость метрики $R^2$ на кросс-валидации (`GridSearchCV.cv_results_['mean_test_score']`) от значения параметра регуляризации для различных степеней полиномов в `PolynomialFeatures`.

```python
indexes_1 = np.array([type(i) == type(Ridge()) for i in grid.cv_results_['param_clf']])
indexes_2 = np.array([type(i) == type(MinMaxScaler()) for i in grid.cv_results_['param_scaler']])
indexes = np.logical_and(indexes_1, indexes_2)
scores = grid.cv_results_['mean_test_score'][indexes]
poly_degrees = grid.cv_results_['param_poly__degree'][indexes]

plt.figure(figsize=(10, 10))
plt.title('Линейная регрессия с L2-регуляризацией и нормировкой признаков в [0,1]')
for i in range(1, 5):
    indexes = (poly_degrees == i)
    plt.semilogx(alphas, scores[indexes], label='Степень полиномов {}'.format(i))
plt.xlabel('Параметр регуляризации')
plt.ylabel('R2')
plt.legend()
plt.tight_layout()
```

При каких значениях происходит переобучение? А недообучение? Почему?

__Ответ:__ Переобучение происходит при alpha < 0.1, то есть вклад регуляризации в функции потерь практически не учитывается, модель запоминает только обучающую выборку, поэтому на тестовой даёт плохой результат. При alpha > 1 вклад регуляризации в функции потерь очень большой, большое число весов зануляется, модель становится простой и недостаточно обучается на выборке.
